```python
import pandas
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv('Graph for visualization of Precision.csv')
print(df)

#dptkan result utk sector lain
#('Consumer Products & Services', 'Industrial Products & Services', 'Property',
#'Transportation & Logisticsâ€™,'Finance Services','Plantation') pastu concat
```

### Graph all sector yearly

```python
def graph_year(year_):
    year=year_
    df_all=df[df.Year==year].sort_values(by="Result_precision")
    plt.figure(figsize=(20,10))
    plt.title('Precision {}'.format(year))
    plt.xlabel('Sector')
    plt.ylabel('Value of Precision')
    
    df_resetindex=df_all.set_index('Year',inplace=False)
    print(df_resetindex)

    plt.bar(df_all['Sector'],df_all['Result_precision'],0.6,color='g')
```

```python
graph_year(2020)
```

```python
graph_year(2019)
```

```python
graph_year(2018)
```

```python
graph_year(2017)
```

```python
graph_year(2016)
```

### Graph by sector last 5 years

```python
def graph_sector(sector_):
    sector=sector_
    df_sector=df[df.Sector==sector].sort_values(by="Year")
    plt.figure(figsize=(20,10))
    plt.title('Precision for Sector {}'.format(sector))
    plt.xlabel('Year')
    plt.ylabel('Value of Precision')
    plt.bar(df_sector['Year'],df_sector['Result_precision'],0.5,color='g')
    df_resetindex=df_sector.set_index('Year',inplace=False)
    print(df_resetindex)
```

```python
graph_sector('Construction')
```

```python
graph_sector('Energy')
```

```python
graph_sector('Telecommunications & Media')
```

```python
graph_sector('Utilities')
```
